```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd

url = "https://www.rottentomatoes.com/top/bestofrt/?year=2019"
page = urlopen(url)
soup = BeautifulSoup(page, 'html.parser')
```

```python
print(soup.select("#top_movies_main > div > table td.right"))
```

```python
reviewLists = []
for review in soup.select("#top_movies_main > div > table td.right"):
    reviewLists.append(review.string.strip())
```

```python
ratingLists = []
for rating in soup.select("#top_movies_main > div > table span.tMeterScore"):
    ratingLists.append(rating.string.strip())
```

```python
movieLists = []
countLists = []
count = 0
for movie in soup.select("#top_movies_main > div > table a"):
    count += 1
    
    print(count, movie.string.strip())
    
    countLists.append(count)
    movieLists.append(movie.string.strip())
```

```python
#print(countLists)
#print(movieLists)
print(ratingLists)
```

```python
# countLists 만으로 DataFrame 생성

df = pd.DataFrame(data=countLists, columns=['Rank'])
df['Rating'] = pd.Series(ratingLists)
df['movies'] = pd.Series(movieLists)
df['Review'] = pd.Series(reviewLists)
df
```

```python
# csv로 저장
df.to_csv("./Data/rottenTomatoesLists.csv", sep=",", encoding='utf-8', index=False)
```

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd

url = "https://www.rottentomatoes.com/top/bestofrt/?year=2017"

page = urlopen(url)
soup = BeautifulSoup(page, 'html.parser')

count = 0
countLists = []
movieLists = []
ratingLists = []
reviewLists = []

list = soup.select("#top_movies_main > div > table a")
for i in range(len(soup.select("#top_movies_main > div > table a"))):
    count += 1
    countLists.append(count)
    movieLists.append(soup.select("#top_movies_main > div > table a")[i].string.strip())
    ratingLists.append(soup.select("#top_movies_main > div > table span.tMeterScore")[i].string.strip())
    reviewLists.append(soup.select("#top_movies_main > div > table td.right")[i].string.strip())
    
```

```python
# countLists 만으로 DataFrame 생성

df = pd.DataFrame(data=countLists, columns=['Rank'])
df['Rating'] = pd.Series(ratingLists)
df['movies'] = pd.Series(movieLists)
df['Review'] = pd.Series(reviewLists)
df
```
